# Gradient Descent
#### Batch gradient descent computes the gradient using the whole dataset.
- We move somewhat directly towards an optimum solution, either local or global. Additionally, batch gradient descent, given an annealed learning rate, will eventually find the minimum located in it's basin of attraction.

#### Stochastic gradient descent (SGD) computes the gradient using a single sample.
- SGD works well (Not well, I suppose, but better than batch gradient descent) for error manifolds that have lots of local maxima/minima. In this case, the somewhat noisier gradient calculated using the reduced number of samples tends to jerk the model out of local minima into a region that hopefully is more optimal.
- One benefit of SGD is that it's computationally a whole lot faster. Large datasets often can't be held in RAM, which makes vectorization much less efficient. Rather, each sample or batch of samples must be loaded, worked with, the results stored, and so on. Minibatch SGD, on the other hand, is usually intentionally made small enough to be computationally tractable.
- Usually, this computational advantage is leveraged by performing many more iterations of SGD, making many more steps than conventional batch gradient descent.

```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

### First we will use a created dataset to show illustration and be able to plot 

Generate sample dataset

```python
def generateData(numPoints, bias, variance):
    
    x = np.zeros((numPoints, 2))
    y = np.zeros((numPoints))
    
    for i in range(0, numPoints):
        x[i][0] = 1
        x[i][1] = i
        y[i] = (i + bias) + random.uniform(0, 1) * variance
        
    return x, y

x, y = generateData(100, 25, 10)
```

Visualize the data

```python
# leave out column of ones 
plt.plot(x[:,1], y, 'o')
plt.show()
```

Calculate gradient

J(θ)=12m(Xθ−y)T(Xθ−y)

```python
def gradientDescent(X, y, theta, alpha, m, numIterations):
    
    xTrans = X.transpose()
    
    for i in range(0, numIterations):
        # Calculate the hypothesis h = X * theta
        hypothesis = np.dot(X, theta)
        # Calculate the loss = h - y and maybe the squared cost (loss^2)/2m
        loss = hypothesis - y
        cost = np.sum(loss **2) / (2 * m)
        if i % 1000 == 0:
            print("Iteration %d | Cost: %f" % (i, cost))
        # Calculate the gradient = X' * loss / m
        gradient = np.dot(xTrans, loss) / m
        # Update the parameters theta = theta - alpha * gradient
        theta = theta - alpha * gradient
        
    return theta
```

```python
m, n = np.shape(x)
numIterations= 80000
alpha = 0.0005
theta = np.ones(n)
theta = gradientDescent(x, y, theta, alpha, m, numIterations)
```

```python
plt.plot(x[:,1], y, 'o')
plt.plot(x[:,1], theta[1] * x[:,1] + theta[0], linewidth = 4)
plt.title('Line Intercept: {}, Slope: {}'.format(*theta))
plt.show()
```

```python
# non numpy way
def stepGradient(b_current, m_current, points, learningRate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        b_gradient += -(2/N) * (points[i].y - ((m_current*points[i].x) + b_current))
        m_gradient += -(2/N) * points[i].x * (points[i].y - ((m_current * points[i].x) + b_current))
    new_b = b_current - (learningRate * b_gradient)
    new_m = m_current - (learningRate * m_gradient)
    return [new_b, new_m]
```

```python
def computeGradient(X, y, alpha, numIterations):
    xTrans = x.transpose()
    m, n = X.shape
    theta = np.ones(n)
    history = []
    costs = []

    for i in range(0, numIterations):
        hypothesis = np.dot(X, theta)
        loss = hypothesis - y
        cost = np.sum(loss **2) / (2 * m)
        costs.append(cost)
        gradient = np.dot(xTrans, loss) / m
        theta = theta - alpha * gradient
        history.append(theta)
    
    return theta, history, costs
```

```python
alpha = 0.001
numIterations = 10000
theta, history, costs = computeGradient(x, y, alpha, numIterations)
```

### Now on a real dataset, so lets load it the boston housing prices

```python
from sklearn.datasets import load_boston
import pandas as pd

boston = load_boston()
data = boston.data
target = boston.target

dat = pd.DataFrame(data, columns = boston.feature_names)
dat['PRICE'] = target
dat.head()
```

Simple linear regresion using the number of rooms

```python
plt.plot(dat.RM, target, 'o')
plt.show()
```

Lets run sklearn's linear regression and basic OLS with numpy to get the parameters

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

X = dat.RM.values
X = sm.add_constant(X)
y = dat.PRICE.values

clf = LinearRegression(fit_intercept = True)
clf.fit(X, y)
print 'Slope: {} -- Intercept: {}'.format(clf.coef_[1], clf.intercept_)
```

```python
intercept, slope = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))
print 'Slope: {} -- Intercept: {}'.format(slope, intercept)
```

```python
plt.plot(X[:,1], y, 'o')
plt.plot(X[:,1], X[:,1] * slope + intercept, linewidth = 3)
plt.show()
```

Now compute using gradient descent

```python
def gradientDescent(X, y, alpha, numIterations):
    X_trans = X.transpose()
    m, n = X.shape
    theta = np.ones(n)
    costs = np.zeros(numIterations)
    
    for i in range(numIterations):
        hypothesis = np.dot(X, theta)
        loss = hypothesis - y
        cost = np.sum(loss **2) / (2 * m)
        costs[i] = cost
        gradient = np.dot(X_trans, loss) / m
        theta = theta - alpha * gradient
        
    return theta
```

```python
numIterations = 100000
alpha = 0.01
theta = gradientDescent(X, y, alpha, numIterations)
theta
```

```python
plt.plot(X[:,1], y, 'o')
plt.plot(X[:,1], X[:,1] * theta[1] + theta[0], linewidth = 3)
plt.show()
```
